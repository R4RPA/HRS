Composition of Discrimination Score


Coding : 1 = Almost every day, 2 = At least once a week, 3 = A few times a month, 4 = A few times a year, 5 = Less than once a year, 6 = Never

Scaling : Create an index of discrimination by reverse - coding all items and a veraging the scores across all six items. Set the final score to missing if there are more than three items with missing values.

Inference: Data for discrimination 6 is missing in questionnarrie so average of 5 values can be considered for year 2006 and average of 6 for other years

Reversing the data

```python
df['discrimination1']=7-df['discrimination1']
df['discrimination2']=7-df['discrimination2']
df['discrimination3']=7-df['discrimination3']
df['discrimination4']=7-df['discrimination4']
df['discrimination5']=7-df['discrimination5']
df['discrimination6']=7-df['discrimination6']

```

```python
cond_2006 = df['year_data'] == 2006
nan_columns = ['discrimination1', 'discrimination2', 'discrimination3', 'discrimination4', 'discrimination5']

df['final_discrimination_score'] = np.where(
    cond_2006,
    np.where(
        df[nan_columns].isna().sum(axis=1) > 2,
        np.nan,
        df[nan_columns].fillna(0).sum(axis=1) / (5-df[nan_columns].isna().sum(axis=1))
    ),
    np.where(
        df[nan_columns + ['discrimination6']].isna().sum(axis=1) > 3,
        np.nan,
        df[nan_columns + ['discrimination6']].fillna(0).sum(axis=1) / (6-df[nan_columns].isna().sum(axis=1))
    )
)
```

```python
columns_to_replace =['trauma1','trauma2',	'trauma3',	'trauma4',	'trauma5',	'trauma6',	'trauma7']
df['Count_Missing'] = df[columns_to_replace].isnull().sum(axis=1)
```

```python
import numpy as np

years_to_replace = [2006, 2008, 2010, 2012]

columns_to_replace =['trauma1','trauma2',	'trauma3',	'trauma4',	'trauma5',	'trauma6',	'trauma7']

for year in years_to_replace:
    for column in columns_to_replace:
        # Create a boolean mask to identify the rows for the specified year
        mask = (df['year_data'] == year)

        # Replace NaN values in the specified column for the specified year
        df.loc[mask, column] = df.loc[mask, column].replace(5, 0)

df['trauma_score']=df['trauma1'].fillna(0)+df['trauma2'].fillna(0)+df['trauma3'].fillna(0)+df['trauma4'].fillna(0)+df['trauma5'].fillna(0)+df['trauma6'].fillna(0)+df['trauma7'].fillna(0)
df['trauma_score']=(df['trauma_score']*7)/(7-df['Count_Missing'])
for i in [2014,2016]:
        # Create a boolean mask to identify the rows for the specified year
        mask = (df['year_data'] == i)

        # Replace NaN values in the specified column for the specified year
        df.loc[mask, 'trauma_score'] = np.nan

# NaN values for 2014 and 2016 will remain unchanged
```


# **Statistical test for checking missing values**

Checking for normal distribution

```python
import seaborn as sns
sns.histplot([x for x in df['age'].values if not np.isnan(x)], kde=True, color='blue')
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram Example with Seaborn')
```

```python
import scipy.stats as stats
col=['age','cognition_score','social_cohesion_score','physical_disorder_score','trauma_score','ctrauma_score','final_discrimination_score']
for i in col:
  dagostino_stat, dagostino_p = stats.normaltest([x for x in df[i].values if not np.isnan(x)])
  print(f"D'Agostino and Pearson Test - Statistic: {dagostino_stat}, p-value: {dagostino_p}")
  if dagostino_p > 0.05:
     print("for ",i," Sample looks like it is normally distributed")
  else:
      print("for ",i,"Sample does not look like it is normally distributed")
```

# Kruskal-Wallis test

To check influence of variables and if they are different



```python
cat_col=['race','hispanic','sex','race_ethnicity']
```

```python
df['hispanic'].unique()
```

```python
import scipy.stats as stats
for i in cat_col:
  for j in col:

    # Create a list to store the data for each category
    groups = []

    # Iterate through unique categories in column A
    for category in df[i].unique():
        # Get the data for the current category, excluding missing values
        group_data = df.loc[df[i] == category, j]
        groups.append(group_data)

    # Perform Kruskal-Wallis test
    kruskal_result = stats.kruskal(*groups)

    # Determine if the categories are significantly different
    alpha = 0.05  # Set your desired significance level
    if kruskal_result.pvalue < alpha:
        print("The categories in variable are significantly different in",i,'with respect to data in',j)
    else:
        print("There is no significant difference among the categories in",i,'with respect to data in',j)
```

```python
for i in cat_col:
  for j in col:# Separate the data into two groups: one with missing values in column A and one with non-missing values in column A
    group_missing = df.loc[df[i].isna(), j]
    group_non_missing = df.loc[~df[i].isna(), j]

    # Perform Kruskal-Wallis test
    kruskal_result = stats.kruskal(group_missing, group_non_missing)

    # Determine if the missing data is significantly different from the non-missing data
    alpha = 0.05  # Set your desired significance level
    if kruskal_result.pvalue < alpha:
        print("The missing data in column",i," is significantly different from the non-missing data in column",j)
    else:
        print("There is no significant difference between the missing data in column",i,"and the non-missing data in column",j)
```
